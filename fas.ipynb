```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def root():
    return {"message": "ok"}

@app.post("/predict")
def predict(data: dict):
    value = data.get("value", 0)

    # 간단한 이상탐지
    if value > 10:
        result = "anomaly"
    else:
        result = "normal"

    return {"result": result}
```
